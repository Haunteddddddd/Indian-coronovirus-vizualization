```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns #Подсмотрел эту библиотеку у одного из пользователей
```

```python
df = pd.read_csv("cases.csv") #Загрузим первый датасет
```

```python
df.head() #Выведу перые 5 строчек для наглядности
```

```python
# Извлечем столбец с общим числом случаев
total_cases = df['new_confirmed']
```

```python
sns.set_style('white') #Приукрасим наши графики
sns.set_palette('Spectral', 4)
plt.rcParams['font.size']=12
```

```python
# Построим график
plt.plot(total_cases)
plt.xlabel('Дата')
plt.ylabel('Общее число случаев')
plt.title('Общее число случаев заболевания коронавирусом в Индии')
plt.show()
```

```python
plt.figure(figsize=(12,6))
#Линия показывающая новые случаи заболевания
sns.lineplot(data=df['new_released'], label="New Released", linewidth = 2)
#Линия показывающая новые случаи смерти
sns.lineplot(data=df['new_deceased'], label="New Deceased", linewidth = 2)
#Линия показывающая накопленные случаи заболевания
sns.lineplot(data=df['acc_released'], label="Accumulated Released", linewidth = 2, linestyle='dashed')
#Линия показывающая накопленные случаи смертей
sns.lineplot(data=df['acc_deceased'], label="Accumulated Deceased", linewidth = 2, linestyle='dashed')
plt.title("Ежедневные случаи заболеваний и смертей от Короновируса в марте 2020 года - Индия")
plt.xlabel("Дата")
plt.ylabel("Количество событий")
plt.show()
```

```python
patient = pd.read_csv("patient.csv") #Возьмем датасет с пациентами
```

```python
patient.head()
```

```python
patient_clean = patient.loc[:, ['gender','age','current_state']].dropna() #Возьмем всего 3 столбца и отчистим датасет от пустых значений
```

```python
sns.set_palette('Spectral', 5) #Приукрасим наш графки, изменив цвет
```

```python
#Сделаем небольшую гистограмку
sns.displot(data=patient_clean, x="age", kde=False) #Незнаю почему, но код не хочет работать с кирилицей, поэтому пришлось написать возраст на английском языке - "age"
plt.title("Распределение пациентов по возрасту")
plt.show()
```

```python
patient_clean['gender'].value_counts() #Сначало посчитаем сколько у нас человек каждого гендера
```

```python
# А теперь сделаем пирог, в котором будет наглядно виден процентаж гендеров
label_gender = ['М', 'Ж']
count_gender = [91, 65] 
explode_gender = [0, 0.1]
plt.pie(count_gender, labels = label_gender, explode = explode_gender, shadow = True,
        autopct='%1.0f%%', startangle=45, pctdistance=1.2, labeldistance=1.4)
plt.title("Процентаж людей по гендеру")
plt.legend(frameon=False, bbox_to_anchor=(0.1,0.2))
plt.show()
```
